```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
mushrooms_data = pd.read_csv('https://stepik.org/media/attachments/course/4852/training_mush.csv')
mushrooms_data.head()
```

```python
X = mushrooms_data.drop(['class'], axis=1)
y = mushrooms_data[['class']]
```

```python
clf_rf = RandomForestClassifier(random_state=0)
```

```python
parametres = {'n_estimators': range(10, 50, 10),
              'max_depth': range(1, 12, 2),
              'min_samples_leaf': range(1, 7),
              'min_samples_split': range(2, 9, 2)}
```

```python
grid_search_cv = GridSearchCV(clf_rf, parametres, cv=3)
```

```python
grid_search_cv.fit(X, y)
```

```python
best_clf = grid_search_cv.best_estimator_
```

```python
grid_search_cv.best_params_
```

```python
feature_importance = best_clf.feature_importances_
feature_importance
```

```python
feature_importance_df = pd.DataFrame({'features': list(X),
                                      'feature_importances': feature_importance})
```

```python
feature_importance_df.sort_values('feature_importances', ascending=False)
```

```python
x_test = pd.read_csv('https://stepik.org/media/attachments/course/4852/testing_mush.csv')
```

```python
x_test.head()
```

```python
predicted = best_clf.predict(x_test)
```

```python
print(sum(predicted))
```

```python
from sklearn.metrics import confusion_matrix

y_true = pd.read_csv('testing_y_mush.csv')
conf_mat = confusion_matrix(y_true, predicted)
```

```python
import seaborn as sns

sns.heatmap(conf_mat, annot=True, cmap='Blues')
```

```python
war_data = pd.read_csv('https://stepik.org/media/attachments/course/4852/invasion.csv')
war_data.head()
```

```python
transport = {'transport' :  0,  'fighter' :  1,  'cruiser' : 2}

X_train = war_data.drop(['class'], axis=1)
y_train = war_data['class'].map(transport)
```

```python
parametres = {'n_estimators': range(10, 50, 10),
              'max_depth': range(1, 12, 2),
              'min_samples_leaf': range(1, 7),
              'min_samples_split': range(2, 9, 2)}

clf_rf = RandomForestClassifier()
```

```python
grid_search_cv = GridSearchCV(clf_rf, parametres, cv=3)
```

```python
grid_search_cv.fit(X_train, y_train)
```

```python
X_test = pd.read_csv('https://stepik.org/media/attachments/course/4852/operative_information.csv')
X_test.head()
```

```python
pd.Series(grid_search_cv.best_estimator_.predict(X_test)).value_counts()
```

```python
pd.DataFrame({'features': list(X_test),
              'feature_importances': grid_search_cv.best_estimator_.feature_importances_}).sort_values('feature_importances', ascending=False)
```
